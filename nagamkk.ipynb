```python
data = [
    (1, "Apple", 100),
    (1, "Apple", 100),
    (1, "Apple", 100),
    (1, "Apple", 100),
    (2, "Orange", 150),
    (3, "Banana", 75),
    (3, "Banana", 75),
    (4, "Grapes", 200),
    (4, "Grapes", 200),
    (4, "Grapes", 200),
    (5, "Kiwi", 20)
]

columns = ["ID", "Fruit", "Count"]

df = spark.createDataFrame(data, columns)

# Display the DataFrame and its schema
df.show()
df.printSchema()
```

```python
from pyspark.sql.functions import sum

df_grp = df.groupBy("Fruit").agg(sum("Count").alias("Total_Count"))
df_grp.show()
```
