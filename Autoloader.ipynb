```python
df = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format", "csv")\
            .option("cloudFiles.schemaLocation", "/Volumes/my_practice_project/my_pp_schema/my_pp_volume/destination/checkpoint/")\
            .option("cloudFiles.schemaEvolutionMode","rescue")\
            .load("/Volumes/my_practice_project/my_pp_schema/my_pp_volume/source/")
```

```python
df.writeStream.format("delta")\
        .option("checkpointLocation", "/Volumes/my_practice_project/my_pp_schema/my_pp_volume/destination/checkpoint/")\
        .trigger(once=True)\
        .start("/Volumes/my_practice_project/my_pp_schema/my_pp_volume/destination/data/")
```
